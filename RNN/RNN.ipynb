# RNN Basics

```
import pandas as pd
import numpy as np
import plotly.express as px
```

## Sine Wave

### Creating Sine Wave

```
x = np.linspace(0,50,501)
x
```

```
y = np.sin(x)
y
```

```
px.line(x=x, y=y)
```

```
df = pd.DataFrame(data=y, index=x, columns=['sin'])
```

### Train test split

```
test_percent = 0.1
test_point = np.round(len(df)*test_percent)
test_ind = int(len(df)-test_point)
train = df.iloc[:test_ind,:]
test = df.iloc[test_ind:,:]
```

```
train
```

```
test
```

### Feature Scaling

```
from sklearn.preprocessing import MinMaxScaler
```

```
n_features=1
```

```
scalar = MinMaxScaler()
scalar.fit(train)
scaled_train = scalar.transform(train)
sclaed_test = scalar.transform(test)
```

### Time Series Generator

```
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```
length = 64 #number of input data which will be used to predict next point
batch_size = 1

generator = TimeseriesGenerator(scaled_train, scaled_train, length=length, batch_size=batch_size)
```

## Simple RNN Model
### Creating model

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM
```

```
model = Sequential()

model.add(SimpleRNN(64, input_shape=(length, 1)))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```
model.summary()
```

### Fitting in the Model

```
model.fit_generator(generator, epochs=5)
```

```
losses = pd.DataFrame(model.history.history)
```

```
losses.plot()
```

### Predicting using the model 

```
first_eval_batch = scaled_train[-length:]
first_eval_batch = first_eval_batch.reshape((1,length,n_features))
```

```
pred = model.predict(first_eval_batch)
```

```
np.append(first_eval_batch, [[pred]])
```

```
sclaed_test[0]
```

```
predictions = []
curr_batch = first_eval_batch
for i in range(len(test)):
  curr_pred = model.predict(curr_batch)[0]
  curr_batch = curr_batch[:,1:,:]
  curr_batch = np.append(curr_batch,[[curr_pred]], axis=1)
  predictions.append(curr_pred)
```

```
predictions = scalar.inverse_transform(predictions)
predictions
```

```
test['prediction'] = predictions
```

```
test
```

```
test.plot()
```

## LSTM Model
### Early stopping and validation data

```
from tensorflow.keras.callbacks import EarlyStopping
```

```
early_stop = EarlyStopping(monitor='val_loss', patience=2)
```

```
validation_generator = TimeseriesGenerator(sclaed_test, sclaed_test, length=length, batch_size=1)
```

```
test_percent = 0.2
test_point = np.round(len(df)*test_percent)
test_ind = int(len(df)-test_point)
train = df.iloc[:test_ind,:]
test = df.iloc[test_ind:,:]
```

```
scalar = MinMaxScaler()
scalar.fit(train)
scaled_train = scalar.transform(train)
sclaed_test = scalar.transform(test)
```

```
generator = TimeseriesGenerator(scaled_train, scaled_train, length=length, batch_size=batch_size)
validation_generator = TimeseriesGenerator(sclaed_test, sclaed_test, length=length, batch_size=batch_size)
```

### Creating the Model

```
model = Sequential()

model.add(LSTM(64, input_shape=(length, 1)))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

### Fitting the model

```
model.fit_generator(generator, epochs=20,
                   validation_data=validation_generator,
                   callbacks=[early_stop])
```

```
pd.DataFrame(model.history.history).plot()
```

### Predicting using Model

```
first_eval_batch = scaled_train[-length:]
first_eval_batch = first_eval_batch.reshape((1,length,n_features))
```

```
predictions = []
curr_batch = first_eval_batch
for i in range(len(test)):
  curr_pred = model.predict(curr_batch)[0]
  curr_batch = curr_batch[:,1:,:]
  curr_batch = np.append(curr_batch,[[curr_pred]], axis=1)
  predictions.append(curr_pred)
```

```
predictions = scalar.inverse_transform(predictions)
predictions
```

```
test['prediction'] = predictions
```

```
test.plot()
```
